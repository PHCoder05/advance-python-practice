```python
import numpy as np

def normal_dist(x, mean, sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density

mean = 0
sd = 1
x = 1
result = normal_dist(x, mean, sd)
print(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Parameters
mean = 0
std_dev = 1
sample_size = 1000

# Generate normal distribution data
data_normal = np.random.normal(mean, std_dev, sample_size)

# Calculate kurtosis
kurt_normal = kurtosis(data_normal)

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.axvline(data_normal.mean(), color='k', linestyle='dashed', linewidth=1, label='Mean')
plt.title(f'Mesokurtic Distribution (Kurtosis: {kurt_normal:.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Parameters
sample_size = 1000

# Generate leptokurtic (Laplace) distribution data
data_leptokurtic = np.random.laplace(0, 1, sample_size)

# Calculate kurtosis
kurt_leptokurtic = kurtosis(data_leptokurtic)

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(data_leptokurtic, bins=30, density=True, alpha=0.6, color='orange', edgecolor='black')
plt.axvline(data_leptokurtic.mean(), color='k', linestyle='dashed', linewidth=1, label='Mean')
plt.title(f'Leptokurtic Distribution (Kurtosis: {kurt_leptokurtic:.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Parameters
sample_size = 1000

# Generate platykurtic (Uniform) distribution data
data_platykurtic = np.random.uniform(-2, 2, sample_size)

# Calculate kurtosis
kurt_platykurtic = kurtosis(data_platykurtic)

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(data_platykurtic, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')
plt.axvline(data_platykurtic.mean(), color='k', linestyle='dashed', linewidth=1, label='Mean')
plt.title(f'Platykurtic Distribution (Kurtosis: {kurt_platykurtic:.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Set up parameters
sample_size = 1000
x = np.linspace(-5, 5, 1000)

# Generate distributions
data_normal = np.random.normal(0, 1, sample_size)       # Mesokurtic
data_leptokurtic = np.random.laplace(0, 1, sample_size)  # Leptokurtic
data_platykurtic = np.random.uniform(-2, 2, sample_size) # Platykurtic

# Calculate kurtosis
kurt_normal = kurtosis(data_normal)
kurt_leptokurtic = kurtosis(data_leptokurtic)
kurt_platykurtic = kurtosis(data_platykurtic)

# Plotting
plt.figure(figsize=(15, 6))

# Normal Distribution (Mesokurtic)
plt.subplot(1, 3, 1)
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.title(f'Mesokurtic (Kurtosis: {kurt_normal:.2f})')
plt.axvline(data_normal.mean(), color='k', linestyle='dashed', linewidth=1)

# Leptokurtic Distribution
plt.subplot(1, 3, 2)
plt.hist(data_leptokurtic, bins=30, density=True, alpha=0.6, color='orange', edgecolor='black')
plt.title(f'Leptokurtic (Kurtosis: {kurt_leptokurtic:.2f})')
plt.axvline(data_leptokurtic.mean(), color='k', linestyle='dashed', linewidth=1)

# Platykurtic Distribution
plt.subplot(1, 3, 3)
plt.hist(data_platykurtic, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')
plt.title(f'Platykurtic (Kurtosis: {kurt_platykurtic:.2f})')
plt.axvline(data_platykurtic.mean(), color='k', linestyle='dashed', linewidth=1)

# Show plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Parameters
mean = 0
std_dev = 1
sample_size = 1000

# Generate normal distribution data
data_normal = np.random.normal(mean, std_dev, sample_size)

# Calculate skewness
skew_normal = skew(data_normal)

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.axvline(data_normal.mean(), color='k', linestyle='dashed', linewidth=1, label='Mean')
plt.title(f'Symmetric Distribution (Skewness: {skew_normal:.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Parameters
sample_size = 1000

# Generate positively skewed (Exponential) distribution data
data_skewed_pos = np.random.exponential(scale=1, size=sample_size)

# Calculate skewness
skew_pos = skew(data_skewed_pos)

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(data_skewed_pos, bins=30, density=True, alpha=0.6, color='orange', edgecolor='black')
plt.axvline(data_skewed_pos.mean(), color='k', linestyle='dashed', linewidth=1, label='Mean')
plt.title(f'Positively Skewed Distribution (Skewness: {skew_pos:.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Parameters
sample_size = 1000

# Generate negatively skewed (Inverse Exponential) distribution data
data_skewed_neg = -np.random.exponential(scale=1, size=sample_size)

# Calculate skewness
skew_neg = skew(data_skewed_neg)

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(data_skewed_neg, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')
plt.axvline(data_skewed_neg.mean(), color='k', linestyle='dashed', linewidth=1, label='Mean')
plt.title(f'Negatively Skewed Distribution (Skewness: {skew_neg:.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```
