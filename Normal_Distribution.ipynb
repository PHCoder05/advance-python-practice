```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 0    # Mean
std_dev = 1 # Standard deviation

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

y = norm.pdf(x, mean, std_dev)

# Plotting the curve
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Normal Distribution\n$\mu={mean}, \sigma={std_dev}$', color='blue')
plt.title('Normal Distribution Curve')
plt.xlabel('X-axis')
plt.ylabel('Probability Density Function (PDF)')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 0    # Mean
std_dev = 1 # Standard deviation
num_samples = 1000  # Number of samples

data = np.random.normal(mean, std_dev, num_samples)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
plt.plot(x, y, color='blue', linewidth=2)

plt.title('Histogram of Normally Distributed Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, norm

data = np.random.normal(0, 1, 1000)  # Normal data
data = np.concatenate([data, np.random.normal(5, 1, 200)])  # Adding a skew

skewness = skew(data)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

x = np.linspace(-4, 10, 1000)
plt.plot(x, norm.pdf(x, np.mean(data), np.std(data)), color='blue')

plt.title(f'Histogram with Skewness: {skewness:.2f}')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, norm

data = np.random.normal(0, 1, 1000)  # Normal data
data = np.concatenate([data, np.random.normal(5, 1, 200)])  # Adding heavy tails

kurt = kurtosis(data)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

x = np.linspace(-4, 10, 1000)
plt.plot(x, norm.pdf(x, np.mean(data), np.std(data)), color='blue')

plt.title(f'Histogram with Kurtosis: {kurt:.2f}')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating a series of data of in range of 1-50.
x = np.linspace(1,50,200)

#Creating a Function.
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density

#Calculate mean and Standard deviation.
mean = np.mean(x)
sd = np.std(x)

#Apply function to the data.
pdf = normal_dist(x,mean,sd)

#Plotting the Results
plt.plot(x,pdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```

```python
# Import required libraries
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

# Creating the distribution
data = np.arange(1, 10, 0.01)
pdf = norm.pdf(data, loc=5.3, scale=1)

# Visualizing the distribution
sb.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sb.lineplot(x=data, y=pdf, color='black')  # Specify x and y parameters
plt.title('Normal Distribution (=5.3, =1)')
plt.xlabel('Heights')
plt.ylabel('Probability Density')
plt.xlim(1, 10)  # Adjust x-axis limits if needed
plt.ylim(0, max(pdf) * 1.1)  # Adjust y-axis limits for better visibility
plt.show()
```
