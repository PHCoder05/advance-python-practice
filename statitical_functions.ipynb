```python
import numpy as np
data = [5,10,15,20,25]
output = np.mean(data)
print(output)
```

```python
import numpy as np
data = [5,10,15,20,25]
output = np.var(data)
print(output)
```

```python
import numpy as np
data = [5,10,15,20,25]
output = np.std(data)
print(output)
```

```python
import numpy as np

data = [2, 4, 6, 8, 10]
data_y = [1, 3, 5, 7, 9]
covariance_matrix = np.cov(data, data_y)
covariance = covariance_matrix[0][1]
print("Covariance:", covariance)
```

```python
import numpy as np

data = [2,4,6,8,10]
data_y = [1,3,5,7,9]
correlation_matrix = np.corrcoef(data, data_y)
correlation = correlation_matrix[0][1]
print("Correlation :", correlation)
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
mean = np.mean(data)

plt.plot(data, label='Data Points')
plt.axhline(y=mean, color='r', linestyle='-', label='Mean')
plt.title('Mean of Data Points')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
mean = np.mean(data)
std_deviation = np.std(data)

plt.hist(data, bins=5, alpha=0.7)
plt.axvline(mean, color='r', linestyle='-', label='Mean')
plt.axvspan(mean - std_deviation, mean + std_deviation, alpha=0.2, color='blue', label='1 SD')
plt.title('Data Spread with Standard Deviation')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
data_y = [1, 3, 5, 7, 9]
correlation = np.corrcoef(data, data_y)[0][1]

plt.scatter(data, data_y)
plt.title(f'Scatter Plot (Correlation: {correlation})')
plt.show()
```
