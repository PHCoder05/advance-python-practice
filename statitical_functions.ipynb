```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

mean = np.mean(data)
median = np.median(data)

plt.hist(data, bins=30, density=True, alpha=0.7, color='gray', edgecolor='black', linewidth=1.2)

plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='blue', linestyle='dashed', linewidth=1.5, label=f'Median: {median:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

plt.legend()

plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(np.arange(-4, 4, 1))
plt.yticks(np.linspace(0, 0.5, 6))

plt.text(mean, 0.4, f'Mean: {mean:.2f}', color='red', ha='center', va='bottom', fontsize=10)
plt.text(median, 0.4, f'Median: {median:.2f}', color='blue', ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def plot_data_and_statistics(data):
    """
    Plots the data with its mean and median, and displays statistical details.

    Parameters:
    data (array-like): The input data to be analyzed and plotted.

    Returns:
    None
    """

    # Calculate statistics
    mean = np.mean(data)
    median = np.median(data)
    std_dev = np.std(data)

    # Print statistics
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")

    # Plot data
    plt.figure(figsize=(10, 6))

    # Plot the data as a histogram
    plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black', density=True)

    # Plot the mean and median lines
    plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean:.2f}')
    plt.axvline(median, color='green', linestyle='dashed', linewidth=1.5, label=f'Median: {median:.2f}')

    # Add title and labels
    plt.title('Histogram of Data with Mean and Median')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()

    # Show plot
    plt.show()

# Example usage
if __name__ == "__main__":
    # Generate random data
    data = np.random.normal(loc=0, scale=1, size=1000)

    # Call the function
    plot_data_and_statistics(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

def scatter_plot_with_regression(x, y):
    """
    Plots a scatter plot of x vs. y with a linear regression line.

    Parameters:
    x (array-like): The x values of the data.
    y (array-like): The y values of the data.

    Returns:
    None
    """
    # Compute linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)

    # Create scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', label='Data points')

    # Plot regression line
    regression_line = slope * x + intercept
    plt.plot(x, regression_line, color='red', linestyle='--', label=f'Regression line\n$y = {slope:.2f}x + {intercept:.2f}$')

    # Add title and labels
    plt.title('Scatter Plot with Linear Regression Line')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()

    # Show plot
    plt.show()

# Example usage
if __name__ == "__main__":
    # Generate random data
    np.random.seed(0)  # For reproducibility
    x = np.random.rand(50)
    y = 2 * x + np.random.normal(scale=0.1, size=x.shape)

    # Call the function
    scatter_plot_with_regression(x, y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_line_with_mean(data):
    """
    Plots a line plot of the data with the mean value annotated.

    Parameters:
    data (array-like): The input data to be plotted.

    Returns:
    None
    """
    # Calculate mean
    mean_value = np.mean(data)

    # Create line plot
    plt.figure(figsize=(8, 5))
    plt.plot(data, marker='o', linestyle='-', color='b', label='Data')

    # Annotate mean
    plt.axhline(mean_value, color='r', linestyle='--', label=f'Mean: {mean_value:.2f}')
    plt.text(len(data) - 1, mean_value, f'Mean: {mean_value:.2f}', color='r', verticalalignment='bottom')

    # Add title and labels
    plt.title('Line Plot with Mean Annotation')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()

    # Show plot
    plt.show()

# Example usage
if __name__ == "__main__":
    # Generate simple data
    data = [2, 3, 5, 7, 11, 13, 17]

    # Call the function
    plot_line_with_mean(data)
```
