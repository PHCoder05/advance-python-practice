```python
import numpy as np

data = [2, 4, 6, 8, 10]
data_y = [1, 3, 5, 7, 9]
covariance_matrix = np.cov(data, data_y)
covariance = covariance_matrix[0][1]
print("Covariance:", covariance)
```

```python
import numpy as np

data = [2,4,6,8,10]
data_y = [1,3,5,7,9]
correlation_matrix = np.corrcoef(data, data_y)
correlation = correlation_matrix[0][1]
print("Correlation :", correlation)
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
mean = np.mean(data)

plt.plot(data, label='Data Points')
plt.axhline(y=mean, color='r', linestyle='-', label='Mean')
plt.title('Mean of Data Points')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
mean = np.mean(data)
std_deviation = np.std(data)

plt.hist(data, bins=5, alpha=0.7)
plt.axvline(mean, color='r', linestyle='-', label='Mean')
plt.axvspan(mean - std_deviation, mean + std_deviation, alpha=0.2, color='blue', label='1 SD')
plt.title('Data Spread with Standard Deviation')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
data_y = [1, 3, 5, 7, 9]
correlation = np.corrcoef(data, data_y)[0][1]

plt.scatter(data, data_y)
plt.title(f'Scatter Plot (Correlation: {correlation})')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Mean and Standard Deviation')

mean = np.mean(data)
std = np.std(data)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(mean + std, color='g', linestyle='dashed', linewidth=2, label='Standard Deviation')
plt.axvline(mean - std, color='g', linestyle='dashed', linewidth=2)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 25, 10, 20]

total = sum(values)
proportions = [value / total for value in values]

plt.pie(proportions, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Distribution by Category')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

mean = np.mean(data)
std_deviation = np.std(data)

plt.hist(data, bins=30, density=True, alpha=0.7, color='gray', edgecolor='black')

plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(mean + std_deviation, color='g', linestyle='dashed', linewidth=2, label='Mean + 1 SD')
plt.axvline(mean - std_deviation, color='g', linestyle='dashed', linewidth=2, label='Mean - 1 SD')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with Mean and Standard Deviation')

plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

mean = np.mean(data)
median = np.median(data)

plt.hist(data, bins=30, density=True, alpha=0.7, color='gray', edgecolor='black', linewidth=1.2)

plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='blue', linestyle='dashed', linewidth=1.5, label=f'Median: {median:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

plt.legend()

plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(np.arange(-4, 4, 1))
plt.yticks(np.linspace(0, 0.5, 6))

plt.text(mean, 0.4, f'Mean: {mean:.2f}', color='red', ha='center', va='bottom', fontsize=10)
plt.text(median, 0.4, f'Median: {median:.2f}', color='blue', ha='center', va='bottom', fontsize=10)

plt.show()
```
