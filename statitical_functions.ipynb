```python
import numpy as np

data = [2, 4, 6, 8, 10]
data_y = [1, 3, 5, 7, 9]
covariance_matrix = np.cov(data, data_y)
covariance = covariance_matrix[0][1]
print("Covariance:", covariance)
```

```python
import numpy as np

data = [2,4,6,8,10]
data_y = [1,3,5,7,9]
correlation_matrix = np.corrcoef(data, data_y)
correlation = correlation_matrix[0][1]
print("Correlation :", correlation)
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
mean = np.mean(data)

plt.plot(data, label='Data Points')
plt.axhline(y=mean, color='r', linestyle='-', label='Mean')
plt.title('Mean of Data Points')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
mean = np.mean(data)
std_deviation = np.std(data)

plt.hist(data, bins=5, alpha=0.7)
plt.axvline(mean, color='r', linestyle='-', label='Mean')
plt.axvspan(mean - std_deviation, mean + std_deviation, alpha=0.2, color='blue', label='1 SD')
plt.title('Data Spread with Standard Deviation')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [2, 4, 6, 8, 10]
data_y = [1, 3, 5, 7, 9]
correlation = np.corrcoef(data, data_y)[0][1]

plt.scatter(data, data_y)
plt.title(f'Scatter Plot (Correlation: {correlation})')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Mean and Standard Deviation')

mean = np.mean(data)
std = np.std(data)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(mean + std, color='g', linestyle='dashed', linewidth=2, label='Standard Deviation')
plt.axvline(mean - std, color='g', linestyle='dashed', linewidth=2)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(2, 1.5, 100)

data = [data1, data2]
plt.boxplot(data, labels=['Data 1', 'Data 2'])
plt.ylabel('Value')
plt.title('Box Plot Comparison')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 25, 10, 20]

total = sum(values)
proportions = [value / total for value in values]

plt.pie(proportions, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Distribution by Category')
plt.show()
```
